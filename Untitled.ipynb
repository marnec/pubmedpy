```python
from bs4 import BeautifulSoup as bs
import requests
import re
import networkx as nx
from IPython.display import Image
from pathlib import Path
import d3fdgraph
import pandas as pd
import matplotlib.pyplot as plt
import pubmedpy as pm
import json
```

```python
xml_article_doc = requests.get('https://dtd.nlm.nih.gov/publishing/tag-library/2.3/index.html').text
```

```python
hierarchy = {}
edges = set()
for m in re.finditer("(\".+\") : (\".+\")", xml_article_doc):
    tag, taglink = m.groups()
    if all(x not in tag for x in ['att', 'model', '.', 'type', 'elements', ':', 'dtd']):
        tag = tag.replace('"', '<', 1).replace('"', '>')
        docfile = Path('/home/marnec/Documents/Journalstruct/dtd.nlm.nih.gov/publishing/tag-library/2.3/' + taglink.replace('"', ''))
        tagname = tag.strip('<>')
        hierarchy[tagname] = {"parents": [], "children": []}
        if docfile.is_file():
            tagfile = bs(open(docfile).read()) 
            elementcontext = tagfile.find('div', {'class': 'elementcontext'})
            if elementcontext:
                for l in elementcontext.find_all('span', {'class': 'elementtag'}):
                    edges.add((l.text.strip('<>'), tagname))
                    hierarchy[tagname]['parents'].append(l.text.strip('<>'))
            
            modeldesc = tagfile.find('div', {'class': 'modeldesc'})
            if modeldesc:
                bullist = modeldesc.find('ul', {'class': 'bullist'})
                if bullist:
                    hierarchy[tagname]['children'] = [ele.text.strip('<>') for ele in bullist.find_all('span', {'class': 'elementtag'})]

json.dump(hierarchy, open('hierarchy.2.3.json', 'w'))
```

```python
# edges = [('sec', 'p'), ('p', 'fig'), ('fig', 'label'), ('fig', 'caption'), ('fig', 'p'), ('caption', 'p')]
G = nx.DiGraph(list(edges))
```

```python
options = {
    'node_color': 'white',
    'node_size': 10,
    'width': 1,
    'arrowstyle': '-|>',
    'arrowsize': 10,
}
fig, axes = plt.subplots(figsize=(20, 20))
pos = nx.spring_layout(G, k=100/len(G.nodes()))
nx.draw_networkx_nodes(G, pos, node_size = 500)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, arrows=True, arrowstyle='-|>');
```
