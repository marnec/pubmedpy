```python
from bs4 import BeautifulSoup as bs
import requests
import re
import networkx as nx
from IPython.display import Image
from pathlib import Path
import d3fdgraph
import pandas as pd
import matplotlib.pyplot as plt
import pubmedpy as pm
import json
```

```python
xml_article_doc = requests.get('https://dtd.nlm.nih.gov/publishing/tag-library/2.3/index.html').text
```

```python
hierarchy = {}
edges = set()
for m in re.finditer("(\".+\") : (\".+\")", xml_article_doc):
    tag, taglink = m.groups()
    if all(x not in tag for x in ['att', 'model', '.', 'type', 'elements', ':', 'dtd']):
        tag = tag.replace('"', '<', 1).replace('"', '>')
        docfile = Path('/home/marnec/Documents/Journalstruct/dtd.nlm.nih.gov/publishing/tag-library/2.3/' + taglink.replace('"', ''))
        tagname = tag.strip('<>')
        hierarchy[tagname] = {"parents": [], "children": []}
        if docfile.is_file():
            tagfile = bs(open(docfile).read()) 
            elementcontext = tagfile.find('div', {'class': 'elementcontext'})
            if elementcontext:
                for l in elementcontext.find_all('span', {'class': 'elementtag'}):
                    edges.add((l.text.strip('<>'), tagname))
                    hierarchy[tagname]['parents'].append(l.text.strip('<>'))
            
            modeldesc = tagfile.find('div', {'class': 'modeldesc'})
            if modeldesc:
                bullist = modeldesc.find('ul', {'class': 'bullist'})
                if bullist:
                    hierarchy[tagname]['children'] = [ele.text.strip('<>') for ele in bullist.find_all('span', {'class': 'elementtag'})]

json.dump(hierarchy, open('hierarchy.2.3.json', 'w'))
```

```python
options = {
    'node_color': 'white',
    'node_size': 2,
    'width': 1,
    'arrowstyle': '-|>',
    'arrowsize': 30,
}
```

```python
G = nx.DiGraph(list(filter(lambda link: any([e in {'p', 'caption'} for e in link]),  edges)))
fig, axes = plt.subplots(figsize=(12, 12))
pos = nx.spring_layout(G, k=30/len(G.edges))

nx.draw_networkx_edges(G, pos=pos, with_labels=True, node_size=1500, alpha=0.3, arrows=True,
        arrowsize=20, width=2)
# draw white circles over the lines
nx.draw_networkx_nodes(G, pos=pos, with_labels=True, node_size=1500, alpha=1, arrows=True,
        arrowsize=20, width=2, node_color='w')
# draw the nodes as desired
nx.draw_networkx_nodes(G, pos=pos, node_size=1500, alpha=.3, arrows=True,
        arrowsize=20, width=2)
nx.draw_networkx_labels(G, pos=pos)
plt.savefig('data/hierarchy_network.png', dpi=150)
```
